###Har xil ma'lumotlar turlari bilan takrorlanuvchi massiv
##Biz op_dtypesiteratsiya paytida elementlarning ma'lumotlar turini o'zgartirish uchun argumentdan foydalanishimiz va unga kutilgan ma'lumotlar turini berishimiz mumkin.
####NumPy o'z joyida (element massivda joylashgan) elementning ma'lumotlar turini o'zgartirmaydi, shuning uchun bu amalni bajarish uchun boshqa joy kerak, bu qo'shimcha joy bufer deb ataladi va uni yoqish uchun nditer()biz o'tkazamiz flags=['buffered'].

```python
import numpy as np

arr3 = np.array([1, 2, 3])

for x in np.nditer(arr3, flags=['buffered'], op_dtypes=['S']):
  print(x)
```

#NumPy Joining Array
##NumPy qo'shilish massivi

####Birlashma ikki yoki undan ortiq massivlar tarkibini bitta massivga joylashtirishni bildiradi.

####SQL-da biz jadvallarni kalit asosida birlashtiramiz, NumPy-da esa massivlarni o'qlar bo'yicha birlashtiramiz.

####Biz concatenate()funktsiyaga qo'shmoqchi bo'lgan massivlar ketma-ketligini eksa bilan birga o'tkazamiz. Agar o'q aniq o'tkazilmasa, u 0 sifatida qabul qilinadi.

```python
arry4 = np.array([1,2,3])
arry5 = np.array([4,5,6])
arr = np.concatenate((arry4,arry5))  # bu ikkita massivni birlashtirish 
```

```python
arr
```

```python
import numpy as np

arr1 = np.array([[1, 2], [3, 4]])

arr2 = np.array([[5, 6], [7, 8]])

arr = np.concatenate((arr1, arr2), axis=1)

print(arr)
```
