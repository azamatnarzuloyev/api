#Simple LR scikit-learning

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head(10)
```

```python
housing = df[df.district=='Чиланзарский']
housing.head()
```

```python
x = housing['size'].to_numpy()
x
```

```python
y = housing['price'].to_numpy()
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=housing, x='size', y='price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=housing, x='size', y='price',line_kws={"color": "red"})
plt.show()
```

```python
y =y.astype(float)
```

```python
ymean = np.mean(y)
```

```python
x = x.astype(float)
```

```python
xmean= np.mean(x)
```

```python
xmean
```

```python
theta1 = sum((x-xmean)*(y-ymean))/sum((x-xmean)**2)
```

```python
theta1
```

```python
theta0 = ymean - theta1*xmean
```

```python
theta0
```

```python
x_test = housing.sample(10, random_state=42)['size'].to_numpy()

y_test = housing.sample(10, random_state=42)['price'].to_numpy()
```

```python
y_predict = theta0 + theta1*x_test
print(y_predict)
```

```python
MAE = np.sum(np.absolute(y_predict-y_test))/len(y_test)
```

```python
RMSE = np.sqrt(np.sum((y_predict-y_test)**2)/len(y_test))
```

```python
RMSE
```

```python
MAE
```

# Scikit-learn yordamida LR
Keling endi yuqoridagi amallarni scikit-learn yordamida bajaramiz.

Avvalo ma'lumotlarni train va testga bo'lib olamiz.

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.10, random_state=42)
```

```python
from sklearn import linear_model
LR_model = linear_model.LinearRegression()

x_train = np.asanyarray(train_set[['size']])
y_train = np.asanyarray(train_set[['price']])

# Modelni tarbiyalaymiz (train)
LR_model.fit(x_train, y_train)

#theta1 va theta0 koeefisentlarni ajratib olamiz 
theta1 = LR_model.coef_[0][0]
theta0 = LR_model.intercept_[0]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=housing, x='size', y='price')
plt.plot(x_train, theta1*x_train + theta0, '-r')
plt.xlabel("Size")
plt.ylabel("Price")
```

```python
x_test = np.asanyarray(test_set[['size']])
y_test = np.asanyarray(test_set[['price']])
```

```python
y_predict = LR_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

MAE = mean_absolute_error(y_test, y_predict)
RMSE = np.sqrt(mean_squared_error(y_test, y_predict))
print(MAE)
print(RMSE)
```

## `curve_fit`
'curve_fit' yordamida o'zimiz yozgan funksiyalar asosida model qurishimiz mumkin. 

Misol uchun, yuqoridagi ma'lumotlar Logistic regression funksiyasi yordamida aniqroq bashorat qilinishi mumkin.

# $$ \hat{Y} = \frac1{1+e^{-\beta_1(X-\beta_2)}}$$

Keling, logisti regression funksiyasini yozib olamiz:

```python
x_data, y_data = (df["Year"].values, df["Value"].values)
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

```python
from scipy.optimize import curve_fit
beta, pcov = curve_fit(sigmoid, xdata, ydata)

print(f"beta1={beta[0]}\nbeta2={beta[1]}")
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
yhat = sigmoid(x, *beta)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,yhat, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
print("MAE:", mean_absolute_error(yhat,ydata))
```
