```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x='price')
plt.show()
```

```python
df = df[df['price']<200000]
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x='price')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x="size")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x="rooms")
plt.show()
```

```python
df = df[df['rooms']<7]
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x="price", hue='rooms')
plt.show()
```

```python
df1 = df[df['rooms']<4]
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df1, x="price", hue='rooms')
plt.show()
```

```python
df1 = df[df['rooms']>3]
plt.figure(figsize=(16,6))
sns.displot(data=df1, x="price", col='rooms')
plt.show()
```

```python

sns.displot(data=df1, x="price", kind='kde',height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=df1, x="price",  shade=True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df[df['rooms'].isin([2,3,5])], x='price', hue='rooms', shade=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="size", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both') # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```
