```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='price', y='size', s=50)
plt.show()
```

```python
df = df[df['size']<250]
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='price', y='size', s=50)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(data=df, x='price', y='size')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='price', y='size',hue='rooms')
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(data=df, x='price', y='size',hue='district')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.show()
```
